# Big O Test

```python
import numpy as np
import matplotlib.pyplot as plt

def create_data(n):
    return np.random.randint(10, size = n)

A = create_data(100)
print(A)
```

```python
import time

def const(A):
  b = A * A

def linear(A):
  for element in A:
    element = element = 10

def cuadratic(A):
  for i in A:
    for j in A:
      element = j * 4


def cubic(A):
  for i in A:
    for j in A:
      for h in A:
        element = i * j

def cublin(A):
  for i in A:
    for j in A:
      for h in A:
        element = i * j
  for element in A:
    element = element * 10
  for element in A:
    element = element * 20
```

```python
sizes = [1, 10, 100]

const_times = []
linear_times = []
cuadratic_times = []
cubic_times = []
cub_lin_times = []

#Const
for n in sizes:
  A = create_data(n)

  start = time.time()
  const(A)
  end = time.time()

  total_time = end - start
  const_times.append(total_time)
  print("N: ", n, " with time: ", total_time)

#Linear
for n in sizes:
  A = create_data(n)

  start = time.time()
  linear(A)
  end = time.time()

  total_time = end - start
  linear_times.append(total_time)
  print("N: ", n, " with time: ", total_time)

#Cuadratic
for n in sizes:
  A = create_data(n)

  start = time.time()
  cuadratic(A)
  end = time.time()

  total_time = end - start
  cuadratic_times.append(total_time)
  print("N: ", n, " with time: ", total_time)

#Cubic
for n in sizes:
  A = create_data(n)

  start = time.time()
  cubic(A)
  end = time.time()

  total_time = end - start
  cubic_times.append(total_time)

  print("N: ", n, " with time: ", total_time)

#Cubic + Linear
for n in sizes:
  A = create_data(n)

  start = time.time()
  cublin(A)
  end = time.time()

  total_time = end - start
  cub_lin_times.append(total_time)

  print("N: ", n, " with time: ", total_time)
```

```python
import seaborn as sns

sns.set_theme()

plt.scatter(sizes, const_times)
plt.plot(sizes, const_times, c = "g", label = "Constant")
plt.scatter(sizes, linear_times)
plt.plot(sizes, linear_times, c = "b", label = "Linear")
plt.scatter(sizes, cuadratic_times)
plt.plot(sizes, cuadratic_times, c = "y", label = "Cuadratic")
plt.scatter(sizes, cubic_times)
plt.plot(sizes, cubic_times, c = "r", label = "Cubic")

plt.xlabel("N")
plt.ylabel("Time in seconds")
plt.title("Functions")
plt.yscale("log")
plt.legend()
```
